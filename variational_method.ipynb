# 変分法

変分法は評価関数を最適化する関数を求めるための手法

ラグランジュの未定乗数法は、目的関数 $f$ を制約条件 $g$ のもとで一次変化($\nabla f$) がゼロとなる停留値(座標)を求める手法に対し、<br>
変分法は評価関数 $J$ の一次変化 $\delta J$ がゼロとなる停留関数を求める。


## ラグランジュの未定乗数法と変分法の対比

### 停留点と停留関数

ラグランジュの未定乗数法では、目的関数 $f$ の一次変化がゼロであることを確認する。<br>
$f$が1変数 $x$ の場合、この一次変化はテイラー展開の $df/dx$ のことである。

$$
f(x + \Delta x) = f(x) + \frac{df}{dx} \Delta x + O(\Delta x^2)
$$

これを図で描くと、以下になる。図では緑の線が$df/dx$を示している。<br>
$x$を変化させていくと、どこかで$df/dx$がゼロになる。そこが停留点である。

<img src="images/taylor_expansion_image.png" style="width:50%;">

変分法では関数の形を変化させることを考える。<br>
これは、関数 $y(x)$ を $\varepsilon \ \eta(x)$ でわずかに変化させる。<br>
始点($x_b$)と終点($x_f$)では$\eta(x_b) = \eta(x_f) = 0$ という条件のもと、$y(x) + \varepsilon \eta(x)$ とすると、<br>
関数は緑のラインのように元の$y(x)$からわずかに変化する。<br>
<img src="images/yx_variation.png" style="width:35%;">

ここで以下のようにとられるようにする。
- $\eta(x)$ : 変分方向
- $\delta y(x) = \varepsilon \eta(x)$  : 微小変分

変分法は停留関数を求めるが、この停留関数は$\delta J=0$という評価関数 $J$の一次変化で確認する。<br>
この$\delta J$は評価関数を次のようなテイラー展開の形で求まる。

$$
J[y(x) + \varepsilon \eta(x)] = J[y(x)] + \delta J \ \varepsilon + O(\varepsilon^2)
$$

これをテイラー展開と同じように図で描く。緑の傾きが$\delta J$であり、これが$J$の一次変化率を示している。
また横軸は関数空間であり、少し位置がずれると数式の形が変わる。例えば $y=sin(x)$や$y=x^2$ のように関数 $y(x)$ が変化する。
<img src="images/J_y_variation_image.png" style="width:70%;">

変分法は候補となる関数$y(x)$に対して、任意の$\eta(x)$ で$\delta J =0$となることで、$y(x)$を停留関数であると判断する。<br>
上図では、任意のではなく、ある一つの$\eta(x)$に対する図になっていることに注意すること。

また、$\varepsilon \ll 1$ のため、評価関数$J[y + \varepsilon \eta(x)]$は$\varepsilon \delta J$ により一次近似ができる。

#### 条件式

ラグランジュの未定乗数法では停留点であるための条件は $\nabla f + \lambda \nabla g = 0$ である。

これと同様の条件式は変分法では、オイラー・ラグランジュ方程式である。

$$
\frac{\partial F}{\partial y} - \frac{d}{dx}\left( \frac{\partial F}{\partial y'}\right) = 0
$$

変分法では評価関数 $J$ の一次変化 $\delta J$ がゼロとなる関数 $y$を探すが、これはあらゆる可能性がある。<br>
$y=x^2$や$y=sin(x)$、または$y=x-3$などである。そのそれぞれの関数に対し、任意の$\eta(x)$で$\delta J=0$ となる停留関数を探すのは現実的ではない。<br>
そこで、オイラー・ラグランジュ方程式で停留関数の形を探す。

#### オイラー・ラグランジュ方程式の導出

変分法では評価関数 $J[y]$ を次の $F$を用いて次のように表す。

$$
J[y] = \int_{x_b}^{x_f} F(x, y, y') dx
$$

$F$ の変数は次のものである。
- $x$ : 位置(独立変数)
- $y$ : 独立変数$x$における関数の値
- $y'$ : 傾き $dy/dx$ 

$x$は通常は1変数の独立変数であり、このノートでは曲線$y(x)$を対象とする変分法を扱う。<br>
一般的に独立変数が $x = [x_0, x_1, ...]$のような多変数へ拡張できるが、<br>
それは場の理論や偏微分方程式を扱う変分法へと発展するためこのノートでは扱わない。

ここで普通の2変数の関数の$f(x,y)$のテイラー展開を考える。

$$
f(x + \Delta x, y + \Delta y) = f(x,y) + \frac{\partial f}{\partial x} \Delta x + \frac{\partial f}{\partial y}\Delta y + O([\Delta x, \Delta y][\Delta x, \Delta y]^T)
$$

変分法では関数 $y(x)$ を $y(x) + \varepsilon \eta(x)$ へ変化させる。<br>
すると、その傾き $y'$も $y' + \varepsilon \eta^{\prime}(x)$ になる。<br>
よって$F(x,y,y')$は$F(x, y + \varepsilon \eta, y' + \varepsilon \eta^{\prime})$となる。<br>
これは、微小変化が$y$と$y'$にしか生じていないため、上記の2変数のテイラー展開と同じになる。

$$
F(x, y + \varepsilon \eta, y' + \varepsilon \eta^{\prime}) = F(x, y, y') + \varepsilon \frac{\partial F}{\partial y}\eta + \varepsilon \frac{\partial F}{\partial y'}\eta^{\prime} + O(\varepsilon ^2)
$$ 

評価関数$J[y]$は以下である。

$$
J[y] = \int_{x_b}^{x_f} F(x, y, y') dx
$$

ここから、$J[y + \varepsilon \eta]$は次のようになる。

$$
J[y + \varepsilon \eta] = \int_{x_b}^{x_f} \left(F(x, y, y') + \varepsilon \frac{\partial F}{\partial y}\eta + \varepsilon \frac{\partial F}{\partial y'} \eta^{\prime}+ O(\varepsilon ^2) \right) dx
$$

積分を分けると、以下のようになる。

$$
J[y + \varepsilon \eta] =  J[y] + \varepsilon\int_{x_b}^{x_f} \left( \frac{\partial F}{\partial y}\eta + \frac{\partial F}{\partial y'} \eta^{\prime} \right) dx + ... 
$$

$$
J[y(x) + \varepsilon \eta(x)] = J[y(x)] + \delta J \ \varepsilon + O(\varepsilon^2)
$$

であるから $\delta J$ は以下となる。

$$
\delta J = \int_{x_b}^{x_f} \left( \frac{\partial F}{\partial y}\eta + \frac{\partial F}{\partial y'}\eta^{\prime}\right) dx
$$

$\delta J$は$\eta$と$\eta^{\prime}$があるため、部分積分で$\eta^{\prime}$を消す。

部分積分は以下であるため、
$$
(f(x)g(x))' = f'(x)g(x) + f(x) g'(x) \rightarrow \int f(x) g'(x) dx = f(x) g(x) - \int f'(x)g(x) dx
$$

$f(x) = \partial F / \partial y' , g'(x) = \eta' $とすると、

$$
\int_{x_b}^{x_f} \frac{\partial F}{\partial y'}\eta^{\prime} dx = \left[ \frac{\partial F}{\partial y'} \eta \right]_{x_b}^{x_f} - \int_{x_b}^{x_f} \frac{d}{dx}\left( \frac{\partial F}{\partial y'} \right) \eta dx
$$

始点と終点は次の条件であった。<br>
$\eta(x_b) = 0 , \ \eta(x_f) = 0$ <br>
そのため上式は次になる。

$$
\int_{x_b}^{x_f} \frac{\partial F}{\partial y'}\eta^{\prime} dx = -\int_{x_b}^{x_f} \frac{d}{dx} \left(\frac{\partial F}{\partial y'} \right)\eta dx
$$

この結果を$\delta J$に戻すと、以下の式になる。

$$
\delta J = \int_{x_b}^{x_f} \left( \frac{\partial F}{\partial y} - \frac{d}{dx} \left(\frac{\partial F}{\partial y'}\right) \right) \eta dx
$$

変分法では、任意の$\eta$を用いて$\delta J=0$となる停留関数を探る。

もし、
$$
\frac{\partial F}{\partial y} - \frac{d}{dx}\left(\frac{\partial F}{\partial y'}\right) \neq 0
$$
となる場所があるなら、その場所で$\eta \neq 0$となる変分も選ぶことができるため、$\delta J \neq 0$ となる。<br>
したがって、任意の$\eta$に対して$\delta J = 0$ とするには、係数そのものが 0 でなければならない。

よって、以下の式が$\delta J = 0$ となる条件である。
$$
\frac{\partial F}{\partial y} - \frac{d}{dx} \left(\frac{\partial F}{\partial y'}\right) = 0
$$

これがオイラー・ラグランジュ方程式である。


## 例題

評価関数を次のように設定する。

$$
J[y] = \int_0^1 F(x,y,y') dx = \int_0^1 (y')^2 dx
$$

$F=(y')^2$は$y'$に関して下に凸であり、傾きが小さいほど評価が低い、という評価関数である。

オイラー・ラグランジュ方程式に当てはめる。

$\partial F / \partial y = 0$, $\partial F / \partial y' = 2 y'$, $d/dx (\partial F / \partial y') = 2 y''$

となるため、オイラー・ラグランジュ方程式より$y''=0$となる。したがって、停留関数は $y(x) = ax + b$である。

## 制御問題へのつながり

変分法では、評価関数

$$
J[y] = \int_{x_b}^{x_f} F(x,y,y') dx
$$
を停留させる関数 y(x) を求めた。

一方、最適制御では求めたいものは曲線 $y(x)$ではなく、時間とともに変化する状態軌道 $x(t)$と制御入力 $u(t)$である。

評価関数も
$$
J = \Phi(x(T)) + \int_0^T L(x,u)dt
$$
のように与えられる。

しかし、状態軌道は自由に選べるのではなく、
$$
\dot{x} = f(x,u)
$$
という運動方程式を満たさなければならない。

そのため、最適制御では
- 評価関数を停留させる
- 運動方程式を制約条件として満たす
という問題になる。

この制約条件をラグランジュの未定乗数法の考えで評価関数へ組み込むことで、ハミルトニアン、随伴方程式、最適性条件が導かれる。

これらをまとめたものがポントリャーギンの最小原理であり、C/GMRESはその最適性条件をリアルタイムに解くアルゴリズムである。

変分法は「関数を最適化する」という考え方を学ぶための基礎であり、最適制御はその考え方を時間発展するシステムへ拡張したものである。
