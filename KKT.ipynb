# KKT条件について

目的関数を制約条件のもと、最小化または最大化する問題を解く条件をKKT条件という。

- 制約条件が等式の場合、ラグランジュの未定乗数法で解く
- 制約条件が `不等式` の場合、KKT条件のもとで解く

## KKT条件による最適値探索問題の概略

目的関数を $f(x, y)$ とし、制約条件 を $g(x,y) \le 0$の不等式条件とし , 随伴変数を$\lambda$ とすると、制約条件 $g$ のもと、目的関数 $f$ の最適値(最小値、または最大値)の候補を探すKKT条件は以下の4つの連立方式式を満たす必要がある。(4)式は相補性条件と呼ばれる。

$$
\begin{array}{ll}
(1) & \nabla f(x,y) + \lambda \nabla g(x,y) = 0 \\
(2) & g(x,y) \le 0 \\
(3) & \lambda \ge = 0 \\
(4) & \lambda g(x,y) = 0
\end{array}
$$






図を例にKKT条件による最適値の探索の例を説明する。

ここで、$\nabla f = [ \partial f / \partial x, \partial f / \partial y]^T = [2x, 2(y-2)]^T$、 $\nabla g = [\partial g / \partial x, \partial g / \partial y] = [0, 1]^T$ である。 $f(x,y)$ は (0,2) を中心にすり鉢状の形をしている。よって、$\nabla f$ は$f$を増大する方向、また、$\nabla g$ は制約領域である $ y \le 1$ とは逆方向、つまり制約を破る方向である。 <br>
<img src ="images/kkt_constrain_image.png" style="width:60%;">

#### case 1

$\nabla f_1$ の出発地点の座標は(0,-1)である。ここでは $g(x,y) = -2 \le 0$ の制約内であり、(2)を満たす。<br>
$g(x,y) = 2$ と (4)の $\lambda g(x,y) = 0$より、$\lambda=0$となり (3)を満たす。<br>
$\lambda=0$ということは、(1)の$\nabla f_1 + \lambda \nabla g = [0, -6]^T + 0 \times [0, 1]^T = [0, -6]^T \neq = 0$ となるため、(1)は満たさない。<br>
よって、この座標は最適解ではないと言える。

この例のように、制約条件の境界であるg(x,y)=0 から離れている場合、(4)の相補性条件より$\lambda=0$ となる。<br>
これは、制約条件の境界を`壁`であると捉えると、座標(0,-1)は壁から離れているため、壁からの反力を受けていない状態であると言える。<br>
これを数式で表現すると $\lambda \nabla g = 0$ となる。 

#### case 2

$\nabla f_2$ の出発地点の座標は(0,1)である。ここでは $g(x,y) = 0 \le 0$ の制約内であり、(2)を満たす。<br>
$g(x,y) = 0$ と (4)の $\lambda g(x,y) = 0$より、$\lambda= any$ である。<br>
$\lambda=any$ということは、(1)の$\nabla f_2 = - \lambda \nabla g = [0, -2]^T = -\lambda \times [0, 1]^T$ より $\lambda = 2$ となるため、(3)と(1)を満たす<br>
よって、この座標は制約条件を加味した最適値の候補であると言える。

この例のように、最適値が制約条件の境界であるg(x,y)=0 にある場合、(4)の相補性条件より$\lambda \neq 0$ となる。<br>
これは、制約条件の境界を`壁`であると捉えると、座標(0,1)は壁に接しており、壁からの反力を受けている状態であると言える。<br>
これを数式で表現すると $\lambda \nabla g \neq 0$ となる。

また、\$nabla f_2 = - lambda \nabla g$ ということは、図からも分かる通り、$\nabla f_2$と $-\lambda \nabla g$が同じ方向を向いており、且つその大きさも一致している、つまり釣り合っている点であると言える。

#### case 3

$\nabla f_3$ の出発地点の座標は(2,1)である。ここでは $g(x,y) = 0 \le 0$ の制約内であり、(2)を満たす。<br>
$g(x,y) = 0$ と (4)の $\lambda g(x,y) = 0$より、$\lambda= any$ である。<br>
$\lambda=any$ということは、(1)の$\nabla f_3 = - \lambda \nabla g = [4, -2]^T = -\lambda \times [0, 1]^T$ となるが、これを満たす $\lambda$は存在しなため、(1)と(3)を満たすことが出来ない。<br>
よって、この座標は最適値ではないと言える。

この例のように、最適値が制約条件の境界であるg(x,y)=0 にあるが、(1)の条件を満たすことができないとき、$\;ambda$は存在しないとなる。<br>

また、\$nabla f_3 \neq - lambda \nabla g$ であることは図からも分かる通り、$\nabla f_3$と $-\lambda \nabla g$が異なる方向を向いているということである。

#### 制約条件の意味

まず、KKT条件では不等式の制約条件を考慮するが、これをスラック変数 $s$ を用いて等式の条件に整える。

例えば、 $g(x,y) = y - 1 \le 0$ という設定とする。ここに$s$という変数を用いると、 $g(x,y) + s^2 = y - 1 + s^2 = 0$ という等式条件とすることが出来る。<br>
スラック変数は必ず正であるという条件を付けることが出来る。<br>
$g(x,y) \le 0$ に $s^2$を付けて、 $g(x,y) + s^2 = 0$ という式を作るとする。<br>
(1) 制約条件を守った場合、 つまり $g(x,g) = -2$ とすると、 $-2 + s^2 = 0 \rightarrow s^2 = 2 \rightarrow s = \sqrt{2}$  <br>
「どんな実数でも、2乗すると0以上になる」というルールから、制約条件を守った場合、実数$s$は存在するとなる。<br>
(2) 制約条件を守らなかった場合、つまり $g(x,y) = 2$ とすると、$2 + s^2 = 0 \rightarrow s^2 = -2$ <br>
となり、2乗してマイナスになる実数は存在しないため、この条件は成立しないとなる。

このように制約条件を守ることを $s^2 \ge 0$ は表裏一体である。

次に、関数の偏微分を取ることを考える。

目的関数の偏微分を $\partial f / \partial x$, $\partial f/ \partial y$ をまとめて $\nabla f$ と記述する。制約関数も同様に $\nabla g$とする。<br>
$\nabla f |_{x,y}$, $\nabla g|_{x,y}$ はそれぞれの $x,y$の位置における傾きを示している。

そして、ある$\lambda$を用いて、その$\nabla f$と$\nabla g$の傾きが反対向きで平行になる $\nabla f = - \lambda \nabla g $ とすることが出来る。<br>
この釣り合った地点が、制約条件を加味した目的関数の最適値(極大,または極小)である。

$\nabla f, \nabla g$は力の大きさであると捉えることが出来る。
ある点でより下の極値の方向動きたいとする目的関数の力(傾き) ($\nabla f$) があり、これに制約条件から押し返すような力($\nabla g$)が発生する。

これをラグランジュ関数 $L(x,y,\lambda) = f(x,y) + \lambda g(x,y)$ と設定して、この偏微分から求める。<br>
この$L(x,y,\lambda)$の制約条件はまだ不等式である。そこでスラック変数 $s^2$を用いて<br>
$L(x,y,\lambda, s) = f(x,y) + \lambda (g(x,y) + s^2)$ として等式の制約条件とする。<br>

偏微分による条件式

$\lambda$は$(x,y)$が決まった後に求まる値である。<br>
$\partial L / \partial x = 0$ , $\partial L / \partial y = 0$ により、$\nabla f$と $\nabla g$ を考慮することができる。<br>
また $\partial L / \partial s = 2 \lambda s = 0 \rightarrow \lambda s = 0 \rightarrow \lambda s^2 = 0$ とする。<br>
そして、スラック変数による等式不等式への変更を行った、 $g(x,y) + s^2 = 0 \rightarrow s^2 = -g (x,y)$を代入すると、<br>
$ \lambda (-g(x,y)) = 0$ の相補性条件となる。

ここから、以下の3式を連立させて、不等式の制約条件を考慮した目的関数の最適値を求める KKT条件を解くことが出来る。

1. $\partial L / \partial x = 0$
1. $\partial L / \partial y = 0$
1. $\lambda(g(x,y)) = 0$

```python
import numpy as np
import scipy
import plotly.graph_objects as go
import sympy
import cvxpy
```

## 目的関数

$$
f(x,y) = x^2 + (y-3)^2
$$

```python
# 目的関数
def f(x,y):
    return x**2 + (y-3)**2
```

```python
# データ作成
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X,Y)

```

## 目的関数の表示

```python
# 目的関数のグラフ作成
fig = go.Figure(
    data=[go.Surface(x=X, y=Y, z=Z)]
)

fig.update_layout(
    title="f(x,y)=x²+(y-3)²",
    scene=dict(
        xaxis_title="x",
        yaxis_title="y",
        zaxis_title="f(x,y)"
    )
)

fig.show()
```

## 制約条件の境界のsurfaceを表示

## 制約条件

$$
g(x,y) = y \le 1
$$

```python
# 制約条件 y <= 1も付けて表示
fig = go.Figure()

# 目的関数
fig.add_surface(
    x=X,
    y=Y,
    z=Z,
    colorscale="Viridis",
    opacity=0.9,
    name="Objective"
)

# 制約面 y = 1
# y = 1 の縦平面
x_plane = np.linspace(np.min(X), np.max(X), 50)
z_plane = np.linspace(np.min(Z), np.max(Z), 50)

X_plane, Z_plane = np.meshgrid(x_plane, z_plane)
Y_plane = np.ones_like(X_plane)

fig.add_surface(
    x=X_plane,
    y=Y_plane,
    z=Z_plane,
    opacity=0.3,
    showscale=False,
    colorscale="Reds",
    name="y = 1"
)
fig.update_layout(title="Objective Function with Constraint y <= 1 with surface")
fig.show()
```

## 境界条件 (y <=1) をマスクして表示

```python
fig = go.Figure()

# 制約条件 y <= 1を満たす目的関数の範囲を表示
Z_mask = np.where(Y <= 1, Z, np.nan)

# 目的関数
fig.add_surface(
    x=X,
    y=Y,
    z=Z_mask,
    colorscale="Viridis",
    opacity=0.9,
    name="Objective"
)

# 制約面 y <= 1 のマスク
x_plane = np.linspace(np.min(X), np.max(X), 50)
z_plane = np.linspace(np.min(Z), np.max(Z), 50)

X_plane, Z_plane = np.meshgrid(x_plane, z_plane)
Y_plane = np.ones_like(X_plane)

fig.add_surface(
    x=X_plane,
    y=Y_plane,
    z=Z_plane,
    opacity=0.3,
    showscale=False,
    colorscale="Reds",
    name="y <= 1"
)

fig.update_layout(title="Objective Function with Constraint y <= 1 with surface (masked)")
fig.show()
```

## ラグランジュ関数L

$$
L(x,y,\lambda) = f(x,x) + \lambda g(x,y)
$$

$f(x,y) = x^2 + (y-3)^2$

$g(x,y) = y \le 1 $ より、 $y -1 \le 0$ とする。

これより、ラグランジュ関数 $L$は

$$
L(x,y,\lambda) = {x^2 + (y-3)^2} + \lambda (y-1)
$$

## 偏微分
$L$ を各変数で偏微分して = 0 とする

$\frac{\partial L}{\partial x} = 2x = 0 \rightarrow x = 0$

$\frac{\partial L}{\partial y} = 2(y-3) + \lambda = 0 \rightarrow \lambda = -2(y-3)$


不等式制約を破らない

$y-1 \le 0$

相補性条件 (壁のスイッチ)

$\lambda (y-1) = 0$

$\lambda \ge 0$ (天井から下向きに「押し返す力 」) 

## ケース分けして計算する

ケースA: もし壁にぶつかっていない ($\lambda =0$ ) としたら?

- $x = 0$
- $2(y - 3)=0 \rightarrow y = 3$

解の候補として、目的関数の低である、$(0,3)$ を得ることが出来る。<br>
しかし制約条件($y \le 1$) を確認すると、
$y = 3$ となっているため制約条件を満たしていない。<br>
天井を突き破っていることになる。よって、この仮定は却下される。


ケースB : 壁にぶつかっている ($\lambda > 0$)と考える

壁にびったり引っ付いている(距離ゼロ)より、 相補条件より $y = 1$

- $x = 0$
- $y = 1$

これを $\partial {L} / \partial {y} $ の式($\lambda = -2(y - 3)$ に代入する。

- $\lambda = -2(1-3) = 4$ 

これは、制約条件($y \le 1$)　と相補条件 $\lambda >0 $ を満たす。

よって、最適解は $(x,y) = (0,1), \lambda=4$ と求まる。
