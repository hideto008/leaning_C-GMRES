# ラグランジュの未定乗数法

目的関数 $f$ を等式制約条件 $g$ のもと、目的関数の`停留点`を求める手法である。
停留点とは、テイラー展開した1階微分または偏微分の項がゼロである点のことである。

一変数の場合のテイラー展開は以下で、停留点では$df/dx = 0$になる。
$$
f(x + \Delta x) = f(x) + \frac{d f}{d x} \Delta x + O(\Delta x ^2)
$$

多変数の場合は以下で、停留点では $\nabla f = \mathbf{0}$ になる。 
$$
f(\mathbf{x} + \Delta \mathbf{x}) = f(\mathbf{x}) + \nabla f^T \Delta \mathbf{x} + O(||\Delta \mathbf{x} ||^2)
$$

$df/dx = 0, \nabla f = \mathbf{0}$ を一次変化がゼロという。<br>
このとき、$f(x+ \Delta x) = f(x) + O( \Delta x ^2)$ と変化は2次以上の小さな値になる。

## ラグランジュの未定乗数法の概略

目的関数を $f(x,y)$ とし、等式制約条件を$g(x,y)=0$とおく。
このとき、以下の関係を満たすとき、その目的関数は制約条件における停留点になる。

$$
\nabla f + \lambda \nabla g = 0
$$

これは制約面上を進む方向に対する$f$の一次変化がゼロであるということを意味している。


## 具体的な数値による例

図では目的関数 $f(x,y) = x^2 + y^2$ で (0,0)を中心としたすり鉢形であり, 制約条件 $g(x,y) = y -2 = 0$ としている。<br> 
また、$\nabla f = [2x, 2y]^T$, $\nabla g = [0, 1]^T$ である。

<img src="images/lagrange_constrain_image.png" style="width:60%;">

制約条件 $g(x,y)$のもとで、目的関数 $f(x,y)$の停留点を求めることを考える。

考えるポイント制約条件$g(x,y)$上の座標であり、ここでいくつかの $\nabla f$ を描くと、図のように $y=2$ の線上から矢印が伸びる。ここで、それぞれの点で $\nabla f + \lambda \nabla g = 0 $ の条件から停留点を求めてみる。 

### case 1

$\nabla f_1$ に関する座標で(0,2)がポイントである。<br>
$\nabla f_1 + \lambda \nabla g = [0,4]^T + \lambda [0,1]^T = 0$ より、$\lambda =-4$ となる。<br>
この$\lambda = -4$より、$\nabla f_1 + \lambda \nabla g = 0$ が満たされるため、これは制約条件上の目的関数の停留点である。


### case 2

$\nabla f_2$ に関する座標で(1,2)がポイントである。<br>
$\nabla f_2 + \lambda \nabla g = [2,4]^T + \lambda [0,1]^T = 0$であり、これを満たす$\lambda$ は存在しない。<br>
よって、この座標は停留点ではない。

### case 3

$\nabla f_3$ に関する座標で(2,2)がポイントである。<br>
$\nabla f_3 + \lambda \nabla g = [4,4]^T + \lambda [0,1]^T = 0$であり、これを満たす$\lambda$ は存在しない。<br>
よって、この座標は停留点ではない。

### 幾何学的な意味

<img src="images/lagrange_constrain_zoom_image.png" style="width:70%;">

図をズームしてもう一度、$\nabla f, \nabla g$の方向を確認する。

まず、$\nabla g$ は $g(x,y) = y-2 = 0 \rightarrow y=2$ という制約になっている。これは 制約面上では$x$ 方向にだけ動ける。よって、図の $\nabla g$ は座標によらず上方向を向いている。

$f(x,y)$ は 制約条件 $g(x,y)$ より $y=2$ の上で $x$ 方向に動くことができる。制約条件におけるそれぞれで場所[(0,2),(1,2),(2,2)] で最も変化する方向 ($\nabla f$) を図で描くと、$\nabla f_1, \nabla f_2, \nabla f_3$ となっている。<br>
$\nabla f = [\partial f / \partial x, \partial f / \partial y] $ である。今 $y=2$ と制約をつけているため、$y$方向へは変化をつけることができない。しかし x 方向には変化することができる。<br>実際、$y=2$として、$\nabla f = [2x, 4]^T$ であり $x$の位置により $\nabla f$ の $x$ 方向が変わるということを意味しており、$\nabla f_1, \nabla f_2, \nabla f_3$ もそのようになっている。<br>

ここでラグランジュの未定乗数法で探したい停留点では制約方向に対して一次変化がない点ということを考えると、変化する余地があるところは停留点ではないということになる。$y$方向は制約で動けないため、考えるべきは$x$方向の一次変化($\partial f/ \partial x)$ である。$\nabla f_2, \nabla f_3$は$x$方向へ直交していないため、この場所では$x$方向に一次変化するよ、と言っているため停留点ではなく、 $\nabla f_1$は$x$方向へ直交しているため、$x$方向には一次変化しないよ、と言っているので停留点である。

偏微分で考えてみても、$\nabla f = [2x, 4]^T$の$x$方向を抜き出して、$x$方向の一次変化がゼロ($\partial f / \partial x =0$) をとってみると、$2x=0 \rightarrow x=0$ となり、これは$\nabla f_1$に関する (0,2)の座標である。

もう一つ他の例で停留点を幾何的に考える。

<img src="images/lagrange_constrain_2_image.png" style="width:65%;">

制約条件の$g(x,y)$は$y$は$x$に対して2次変化している。そのため、制約方向に直交する$\nabla g$は場所により様々な方向を向いている。しかし、制約方向に対しては直交している。

目的関数 $f$ は制約条件 $g$ 下では$\nabla f_1, \nabla f_2, \nabla f_3$ となる。$\nabla f$ の制約方向に対する一次変化がないとは、$\nabla f$が制約方向に対して直交であるということが条件になる。これが停留点である。

$\nabla g$は常に制約方向に対して直交になる。そのため、$\nabla g$と平行な $\nabla f$ が停留点である。$\nabla f_1, \nabla f_2$は$\nabla g$ と平行ではなく、$\nabla f_3$ が $\nabla g$ と平行である。よって$\nabla f_3$の点(0,2)が、制約条件を考慮した目的関数の停留点である。

ただ、$\nabla f$ と $\nabla g$ が平行である、ということを式として条件にするため、ラグランジュの未定乗数 $\lambda$を用い、$\nabla f + \lambda \nabla g =0$ という条件式にしている。

まとめると、<br>
ラグランジュの未定乗数法は、等式制約 $g(x)=0$ のもとで、目的関数 $f(x)$ の制約付き停留点を求める方法である。<br>
制約付き停留点では、制約面に沿った任意の微小変化 $\Delta x$ に対して、一次変化 $\nabla f^T \Delta x$ がゼロになる。<br>
そのため、$\nabla f$ は制約面の接線方向に直交し、制約面の法線である $\nabla g$ と平行になる。これを式で表すと $\nabla f+ \lambda \nabla g=0$ となる。

### 最適制御へのつながり

最適制御では、最適化する対象は点$x$ ではなく、時間とともに変化する状態軌道 $x(t)$ になる。<br>
また制約条件も $\dot{x} = f(x,u)$ の運動方程式になる。

そのため、ラグランジュの未定乗数法の考え方を時間とともに変化する軌道へ拡張したものとして変分法が導入され、ハミルトニアン $H = L + \lambda ^T f$ が定義される。

C/GMRESでは、このハミルトニアンから得られる最適性条件
$$
\frac{\partial H}{\partial u} = 0
$$
をリアルタイムに解き続けることで最適制御入力を求める。

したがって、C/GMRESを理解するためには、
- 停留点とは何か 
    - (テイラー展開の一次微分(偏微分)の項がゼロの点。この一次微分を一次変化と呼ぶ)
- 一次変化をゼロにする意味 
    - (制約面に沿った方向(制約面上を進む方向)への一次変化がゼロとなる点(停留点)を探す)
- ラグランジュ乗数の役割 
    - ($\nabla f$ と $\nabla g$ が平行であるという幾何学的条件を、$\nabla f+\lambda \nabla g=0$ という方程式で表現するための未知数。また、目的関数に対する制約の影響の大きさを表す。)

を理解することが重要となる。
